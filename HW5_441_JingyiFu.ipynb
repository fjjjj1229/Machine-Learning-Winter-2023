# 2.) Choose 4 features you think are important. Case_status is your target variable


```python
df.info()
```

```python
df_select = df[["class_of_admission","application_type","us_economic_sector","wage_offer_unit_of_pay_9089","case_status"]]
```

# 3.) Clean your data for a decision tree


```python
df_select.dropna(inplace=True)
df_select.reset_index(inplace=True)
df_select
```

```python
dummies = pd.get_dummies(df_select[["class_of_admission","application_type","us_economic_sector","wage_offer_unit_of_pay_9089"]])
y = df_select["case_status"]
X = pd.concat([dummies],axis=1)
```

```python
X.head()
```

```python
X.columns
```

# 4.) Fit and plot a decision tree of depth 3


```python
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
```

```python
clf = tree.DecisionTreeClassifier(max_depth = 3)
clf.fit(X,y)
```

```python
plt.figure(figsize = (20,8))
tree.plot_tree(clf,class_names=y,filled=True)
```

# 6.) Using a for loop, make your own train-test split and determine the best max_depth for out-of sample accuracy

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
```

```python
outputs = []
max_depths = [1,2,3]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3)

for md in max_depths:
  clf = DecisionTreeClassifier(max_depth=md)
  clf.fit(X_train, y_train)
  y_pred = clf.predict(X_test)

  # Test the model with out of sample data
  accuracy = accuracy_score(y_test, y_pred)
  outputs.append(accuracy)
print(outputs)
```

```python
print("The outcome shows that the accuracy scores are really close among different depth")
print("It seems that the outputs which depth = 2 is better than others.")
```
